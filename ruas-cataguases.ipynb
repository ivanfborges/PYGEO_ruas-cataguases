```python
import osmnx as ox
import folium
import geopandas as gpd
from geopandas.tools import geocode

# Defina a localização (cidade, estado, país) que você deseja extrair os limites do município
municipio = "Cataguases, Minas Gerais, Brasil"

# Use a função `graph_from_place` para extrair o grafo de ruas da localização
graph = ox.graph_from_place(municipio, network_type='all')

# Salve o shapefile das ruas
filepath=r"D:\Downloads\ruas-cataguases.shp"
ox.save_graph_shapefile(graph, filepath)

# Carregue o shapefile das ruas
ruas = gpd.read_file(r"D:\Downloads\ruas-cataguases.shp\edges.shp")

# Use o osmnx para obter os limites do município
limites_municipio = ox.geocode_to_gdf(municipio)

# Salve o shapefile dos limites do município
limites_municipio.to_file(r"D:\Downloads\ruas-cataguases.shp\limites_municipio.shp")

# Carregue o shapefile dos limites do município
limites = gpd.read_file(r"D:\Downloads\ruas-cataguases.shp\limites_municipio.shp")

# Crie um mapa centrado na localização desejada
mapa = folium.Map(location=[limites.geometry.centroid.y.mean(), limites.geometry.centroid.x.mean()], zoom_start=11)

# Adicione a camada de imagem de satélite
folium.TileLayer('Stamen Terrain').add_to(mapa)

# Adicione a camada dos limites do município
folium.GeoJson(limites.geometry.to_crs(epsg=4326), name='Limites Cataguases/MG', style_function=lambda x: {'color': 'red', 'fillOpacity': 0}).add_to(mapa)

# Adicione a camada das ruas
folium.GeoJson(ruas.geometry.to_crs(epsg=4326), name='Ruas', style_function=lambda x: {'color': 'black', 'weight': 0.5}).add_to(mapa)

# Adicione controle de camadas (layers control)
folium.LayerControl().add_to(mapa)

# Crie uma legenda personalizada
legenda_html = '''
     <div style="position: fixed; 
                 bottom: 50px; left: 50px; width: 180px; height: 70px; 
                 border:2px solid grey; z-index:9999; font-size:14px;
                 background-color: transparent;
                 ">
         <p style="margin:10px"><span style='color:red;'>&#9632;</span> Limites do Município</p>
         <p style="margin:10px"><span style='color:black;'>&#9632;</span> Ruas</p>
     </div>
'''
mapa.get_root().html.add_child(folium.Element(legenda_html))

# Adicione um título ao mapa
titulo_html = '''
     <div style="position: fixed; 
                 top: 50px; left: 50%; transform: translateX(-50%); width: 320px; 
                 text-align: center; z-index:9999; font-size:28px;
                 background-color: transparent;
                 ">
         Ruas de Cataguases
     </div>
'''
mapa.get_root().html.add_child(folium.Element(titulo_html))

# Salve o mapa em um arquivo HTML
mapa.save(r"D:\Downloads\ruas-cataguases.shp\ruas-cataguases.html")

# Exiba o mapa
mapa
```
